```python
# Compare Datasets T0 and T1 --> Height

import pandas as pd
import numpy as np

t0_data = pd.read_csv('T0_With questionaire.csv', sep=';')

t1_data = pd.read_csv('HHS_T1_20_10_21.csv', error_bad_lines=False, sep=';')

t0_data['Lengte 2 (afgerond op 0.1 cm)'] = t1_data['Lengte 2 (afgerond op 0.1 cm)']
t0_data['Height_difference'] = t1_data['Lengte 2 (afgerond op 0.1 cm)'].astype(float) - t0_data['Lengte (afgerond op 0.1 cm)'].astype(float)

count_pos_n = 0
count_neg_n = 0 

for num in t0_data['Height_difference']:
    if num >= 0:
        count_pos_n += 1
    else:
        count_neg_n += 1

print("Count of positive differences: ", count_pos_n) 
print("Count of negative differences: ", count_neg_n)

print(t0_data)
```

```python
t0_data.set_index('Respondentnummer', inplace=True)
t1_data.set_index('Respondentnummer', inplace=True)

print(t0_data['Height_difference'])

print(np.mean(t0_data['Height_difference']))
```
