```python
# Compare Datasets T0 and T1 --> MQ Score

import pandas as pd
import numpy as np

t0_data = pd.read_csv('T0_With questionaire.csv', sep=';')
# print(t0_data)

t1_data = pd.read_csv('HHS_T1_20_10_21.csv', error_bad_lines=False, sep=';')
# print(t1_data)

# t0_data.set_index('Respondentnummer')
# t1_data.set_index('Respondentnummer')
t0_data['MQ score'] = t1_data['MQ score']
# t0_data['MQ_match'] = np.where(t0_data['MQ'] == t1_data['MQ score'], 'True', 'False')
t0_data['MQ_difference'] = t1_data['MQ score'].astype(float) - t0_data['MQ'].astype(float)

count_pos_n = 0
count_neg_n = 0 

for num in t0_data['MQ_difference']:
    if num >= 0:
        count_pos_n += 1
    else:
        count_neg_n += 1

print("Count of positive differences: ", count_pos_n) 
print("Count of negative differences: ", count_neg_n)

print(t0_data)
```
