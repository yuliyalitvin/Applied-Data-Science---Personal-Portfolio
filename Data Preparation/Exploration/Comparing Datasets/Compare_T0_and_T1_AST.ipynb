```python
# Compare Datasets T0 and T1 --> AST

import pandas as pd
import numpy as np

t0_data = pd.read_csv('T0_With questionaire.csv', sep=';')

t1_data = pd.read_csv('HHS_T1_20_10_21.csv', error_bad_lines=False, sep=';')

t0_data['AST-1 (tijd afgerond op 0.1s)'] = t1_data['AST-1 (tijd afgerond op 0.1s)']
t0_data['AST_difference'] = t1_data['AST-1 (tijd afgerond op 0.1s)'].astype(float) - t0_data['AST (tijd afgerond op 0.1s)'].astype(float)

count_pos_n = 0
count_neg_n = 0 

for num in t0_data['AST_difference']:
    if num >= 0:
        count_pos_n += 1
    else:
        count_neg_n += 1

print("Count of positive differences: ", count_pos_n) 
print("Count of negative differences: ", count_neg_n)

print(t0_data)
```

```python
t0_data.set_index('Respondentnummer', inplace=True)
t1_data.set_index('Respondentnummer', inplace=True)

print(t0_data['AST_difference'])

print(np.mean(t0_data['AST_difference']))
```
